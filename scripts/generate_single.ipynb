```python

import gzip
import json
import os

```

```python

# os.mkdir("/hy-tmp/data/datasets/R2R_VLNCE_NRSub/single")
episode_id = 10778
```

```python
split = "train"
target = "single"
sub_data_path = "/home/ubuntu/MLA/data/datasets/R2R_VLNCE_NRSub/%s/%s_sub.json.gz"%(split,split)
with gzip.open(sub_data_path, "r") as f:
    sub_data = json.loads(f.read())
data = None
for v in sub_data["episodes"]:
    if int(v["episode_id"])==episode_id:
        data = v
        break
sub_data["episodes"] = [data]
target_data_path = "/home/ubuntu/MLA/data/datasets/R2R_VLNCE_NRSub/%s/%s_sub.json.gz"%(target,target)
with gzip.open(target_data_path, "w") as f:
    f.write(json.dumps(sub_data).encode("utf-8"))
print(sub_data["episodes"])

sub_data_path = "/home/ubuntu/MLA/data/datasets/R2R_VLNCE_NRSub/%s/%s_gt.json.gz"%(split,split)
with gzip.open(sub_data_path, "r") as f:
    sub_data = json.loads(f.read())
data = {}
for k,v in sub_data.items():
    if int(k)==episode_id:
        data[k] = v
        break
target_data_path = "/home/ubuntu/MLA/data/datasets/R2R_VLNCE_NRSub/%s/%s_gt.json.gz"%(target,target)
with gzip.open(target_data_path, "w") as f:
    f.write(json.dumps(data).encode("utf-8"))
print(data)
```
